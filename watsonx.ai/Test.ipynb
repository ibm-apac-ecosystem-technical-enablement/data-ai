```python
# Import libraries
import json
import os
import pandas as pd

# Set the max column width to None (no truncation)
pd.set_option('display.max_colwidth', None)

from dotenv import load_dotenv
#from ibm_cloud_sdk_core import IAMTokenManager
from ibm_watson_studio_lib import access_project_or_space
#from ibm_cloud_sdk_core.authenticators import IAMAuthenticator

# WML python SDK
from ibm_watson_machine_learning.foundation_models import Model
from ibm_watson_machine_learning.metanames import GenTextParamsMetaNames as GenParams
from ibm_watson_machine_learning.foundation_models.utils.enums import ModelTypes, DecodingMethods
```

```python
# Define connection variables
api_key = "YOUR IBM CLOUD API KEY"
ibm_cloud_url = "https://us-south.ml.cloud.ibm.com"
project_id = "YOUR WATSONX.AI PROJECT ID"
access_token = "OUR WATSONX.AI PROJECT ACCESS TOKEN"
```

```python
prompt = ("What is the capital city of Australia?")
```

```python
model_init = None
# Initialize the watsonx model
def get_model(model_type,max_tokens,min_tokens,decoding,temperature):#, repetition_penalty):

    generate_params = {
        GenParams.MAX_NEW_TOKENS: max_tokens,
        GenParams.MIN_NEW_TOKENS: min_tokens,
        GenParams.DECODING_METHOD: decoding,
        GenParams.TEMPERATURE: temperature,
    }
    global model_init
    if model_init == None:
        model_init = Model(
            model_id=model_type,
            params=generate_params,
            credentials={
                "apikey": api_key,
                "url": ibm_cloud_url
            },
            project_id= project_id
            )

    return model_init
```

```python
# Set up watsonx model and parameters
model_type = "ibm/granite-13b-instruct-v2"
max_tokens = 2000
min_tokens = 5
decoding = DecodingMethods.GREEDY
temperature = 0.7

# Get the watsonx model
model = get_model(model_type, max_tokens, min_tokens, decoding, temperature)
```

```python
# Send a prompt to model
generated_response = model.generate(prompt)
response_text = generated_response['results'][0]['generated_text']
```
